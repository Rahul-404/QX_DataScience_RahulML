```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

```python
# create dataset
np.random.seed(1)
X = np.arange(0, np.pi*2, 0.1).reshape(-1,1)
y = np.sin(X)**3 + np.random.randn(*X.shape)*0.1
```

```python
plt.scatter(X,y,c='darkorange',edgecolors='black')
```

```python
# regression models
model_1 = DecisionTreeRegressor(max_depth=2)
model_2 = DecisionTreeRegressor(max_depth=5)
model_3 = DecisionTreeRegressor(max_depth=8)
model_1.fit(X,y)
model_2.fit(X,y)
model_3.fit(X,y)
```

```python
#predict
x_grid = np.arange(min(X), max(X), 0.01)
x_grid = x_grid.reshape(len(x_grid), 1)
y_1 = model_1.predict(x_grid)
y_2 = model_2.predict(x_grid)
y_3 = model_3.predict(x_grid)
```

```python
# plot
plt.scatter(X,y, edgecolor='black', c='orange')
plt.plot(x_grid,y_1, color='cornflowerblue',
        label='max_depth=2', linewidth=2)
plt.plot(x_grid,y_2, color='yellowgreen',
        label='max_depth=5', linewidth=2)
plt.plot(x_grid,y_3, color='red',
        label='max_depth=3', linewidth=2)
plt.xlabel('Data')
plt.ylabel('Target')
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
print('# max_depth=2 :',r2_score(y, model_1.predict(X)))
print('# max_depth=5 :',r2_score(y, model_2.predict(X)))
print('# max_depth=8 :',r2_score(y, model_3.predict(X)))
```
