```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# np.random.seed(10)
X = np.arange(0, 2*np.pi, 0.5)
y = np.sin(X)**3 + np.random.randn(*X.shape)*0.1
```

```python
plt.scatter(X,y, color='red')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('X vs Y')
plt.show()
```

```python
X.shape
```

```python
X = X.reshape(-1,1)
```

```python
y
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
# svr model
model = SVR()
model.fit(X,y)
# linear model
lr = LinearRegression()
lr.fit(X,y) 
```

```python
print('RMSE of Linear Model: {:.2f}'.format(np.sqrt(mean_squared_error(y, lr.predict(X)))))
print('R2 Score of Linear Model: {:.2f}'.format(r2_score(y, lr.predict(X))))
```

```python
print('RMSE of SVR Model: {:.2f}'.format(np.sqrt(mean_squared_error(y, model.predict(X)))))
print('R2 Score of SVR Model: {:.2f}'.format(r2_score(y, model.predict(X))))
```

```python
x_grid = np.arange(min(X),max(X),0.01)
x_grid = x_grid.reshape(len(x_grid),1)
plt.scatter(X,y, color = 'black',label='Data')
plt.plot(x_grid, model.predict(x_grid), color='red',label='svr')
plt.plot(x_grid, lr.predict(x_grid), color='green',label='Linear model')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
print('R2 Score of SVR: {:.2f}'.format(r2_score(y,model.predict(X))))
print('R2 Score of Linear model: {:.2f}'.format(r2_score(y,lr.predict(X))))
```
